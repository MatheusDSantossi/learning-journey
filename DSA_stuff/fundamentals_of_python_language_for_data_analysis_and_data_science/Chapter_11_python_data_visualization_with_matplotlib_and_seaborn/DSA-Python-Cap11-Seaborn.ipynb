### Creating graphics with Seaborn

```python
# Loading one of the datasets that comes with Seaborn
data = sea.load_dataset("tips")
```

```python
data.head()
```

```python
# The method joinplot creates a plot of 2 variables with bivariates and univariates
sea.jointplot(data, x = "total_bill", y = 'tip', kind = "reg")
```

```python
# The method lmplot() creates a plot with data and regression models
sea.lmplot(data, x = "total_bill", y = "tip", col = "smoker")
```

```python
# Building a dataframe with Pandas
df = pd.DataFrame()
```

```python
# Feeding the dataframe with random values
df['age'] = random.sample(range(20, 100), 30)
df['weight'] = random.sample(range(55, 150), 30)
```

```python
df.shape
```

```python
df.head()
```

```python
# lmplot()
sea.lmplot(data = df, x = "age", y = "weight", fit_reg= True)
```

```python
# kdeplot
sea.kdeplot(df.age)
```

```python
# kdeplot
sea.kdeplot(df.weight)
```

```python
# distplot
sea.distplot(df.weight)
```

```python
# Histogram
plt.hist(df.age, alpha = .3)
sea.rugplot(df.age)
```

```python
# Box Plot
sea.boxplot(df.age, color = 'm')
```

```python
# Box Plot
sea.boxplot(df.weight, color = "y")
```

```python
# Violin Plot
sea.violinplot(df.age, color = "g")
```

```python
# Violin Plot
sea.violinplot(df.weight, color = "cyan")
```

```python
# Clustermap
sea.clustermap(df)
```

### Using Matplotlib, Seaborn, NumPy and Pandas in the Creation of Statistical Graphics

```python
# Random Values
np.random.seed(42)
n = 1000
pct_smoke = 0.2

# Variables
flag_smoke = np.random.rand(n) < pct_smoke
age = np.random.normal(40, 10, n)
height = np.random.normal(170, 10, n)
weight = np.random.normal(70, 10, n)

# Dataframe
data = pd.DataFrame({ 'height': height, "weight": weight, "flag_smoke": flag_smoke})

# Creates the data for the flag_smoke variable
data["flag_smoke"] = data["flag_smoke"].map({True: "Smoke", False: "No Smoke"})
```

```python
data.shape
```

```python
data.head()
```

```python
# Style 
sea.set(style = "ticks")

# lmplot
sea.lmplot(x = "height",
          y = "weight",
          data = data,
          hue = "flag_smoke",
          palette = ["tab:blue", "tab:orange"], 
          height = 7
          )

# Labels and title
plt.xlabel("Height (cm)")
plt.ylabel("Weight (kg)")
plt.title("Relation Between Height and Weight of Smoke No-Smoke People")

# Remove the bords
sea.despine()

# Show
plt.show()
```
