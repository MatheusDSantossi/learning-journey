### Building Plots

In Matplotlib, a "plot" refers to a graphical representation of data, typically showing the relationship between two or more variables. It's the visual output generated by Matplotlib's functions, most commonly using the matplotlib.pyplot.plot() function.

Here's a breakdown of what a plot entails in Matplotlib:

Types of Plots:

* Matplotlib supports a wide array of plot types, each suited for different data visualization needs, including:
* Line Plots: Showing trends over time or continuous variables.
* Scatter Plots: Illustrating the correlation between two variables.
* Bar Charts: Comparing discrete categories.
* Histograms: Displaying the distribution of a single variable.
* Pie Charts: Representing proportions of a whole.
* Contour Plots: Visualizing 3D data on a 2D plane.
* 3D Plots: Representing data in three dimensions.

In essence, a plot in Matplotlib is the final visual output that helps in understanding and communicating insights from data.

### Filling Absent Values in Pandas DataFrames

The function fillna() is used to fill absent values. The function offers many options. We can use a specific value, an aggregate functon (for instance, average() or the previous values or the next.

For that example we'll use the mode, the statistic represents the value that appear more times in a variable

```python
# First we import a dataset
df = pd.read_csv("dataset.csv")
```

```python
df.head(5) # 5 is the default value
```

```python
df.isna().sum()
```

```python
# We extract the mode of Quantity column
mode = df["Quantidade"].value_counts().index[0]
```

In mathematics, the mode is the value that appears most frequently in a set of data. It's a measure of central tendency, like the mean and median, used to describe the typical value in a dataset. 

In data analysis, the mode helps identify the most common occurrence or trend within a dataset, which can be useful for various applications. 

```python
print(mode)
```

```python
# And in the end we fill the values NA with the mode
df["Quantidade"].fillna(value = mode, inplace = True) 
# inplace = True, it tells pandas to replace the values in the dataframe
# itself instead of creating a copy
```

```python
df.isna().sum()
```

### Checking the Occurrence of Several Values in a Column

In our example dataset we have the column Quantidade that represents the quantity of sold items in each one of the sales. Imagine that we need to know in which sales were sold 5, 7, 9 or 11 items.

Will we apply that type of filter in our dataframe?

Easy. Pandas offers the method isin() to check several values in a column. Who knows SQL Language must already have noticed that the method is the same as the clause IN in SQL. Let's go to the example.

```python
df.shape
```

```python
# Then we apply the filter
df[df["Quantidade"].isin([5, 7, 9, 11])]
```

In the instruction above, we're filtering the dataframe called f, returning all the lines where the column Quantity is equal to the values 5, 7, 9 or 11. We give a list of values as argument for the method isin().

Let's make it a little more fun. If you executed the instruction above, you noticed that were returned 2.128 lines. And if we want to return only 10 lines? Just slice the result like this:

```python
# Shape
df[df["Quantidade"].isin([5, 7, 9, 11])].shape
```

```python
df[df["Quantidade"].isin([5, 7, 9, 11])][:10]
```

### Logical Operators for Data Manipulation with Pandas

The Logical Operators are excelent to filter dataframes and return exactly the data we need for our work. To know more about the logical operators rules, go to this page:

https://press.rebus.community/programmingfundamentals/chapter/logical-operators/

FIrst we'll use the logical operator AND to check two conditions. It'll be returned the registers when both of the conditions were true

```python
# FIltering the sales that happened for working from 
# home segment and in the South region
df[ (df.Segmento == "Home Office") & (df.Regiao == "South") ].head()
```

But it might be needed to check two conditions and return the register if at least one of them are true. In that case, we use the operator OR, like we show bellow

```python
# Filtering the sales that happened for working from 
# home or South region
df[(df.Segmento == "Home Office") | (df.Regiao == "South")].tail()
```

The not operator is the opposite of the first example.

```python
# Filtering the sales that neither happened for working from 
# home or South region
df[(df.Segmento != "Home Office") & (df.Regiao != "South")].sample(5)
```

### Grouping Data in DataFrames with Group By

The Pandas function Groupby is a versatile and easy-to-use function that helps to take a general vision of data. That makes it easier to explore the dataset and reveal the relationship between the variables.

The following code will group the lines based in combinations Segmento/Regiao/Valor_Venda, and it will give the sales average rate of each group.

```python
# Applying group by
df[["Segmento", "Regiao", "Valor_Venda"]].groupby(["Segmento", "Regiao"]).mean()
```

In the above instruction we filter the data extracting 3 columns: ['Segmento', 'Region', 'Valor_Venda']. In the sequence, we group for two columns: ['Segmento', 'Regiao]. And then we calculate the average for the column that were from group by, in that case the column Sales.

The behavior of group by with Pandas is the same observed in SQL Language

### Multiple Agregation with Group By

Let's explore more the groupby() function  because we have several options of simple data summarization. In the example below, we'll unify the function groupby() with the function agg() that realizes multiple aggregation

```python
# Apply group by
df[["Segmento", "Regiao", "Valor_Venda"]].groupby(["Segmento", "Regiao"]).agg("mean", "std", "count")
```

In the instruction above, first we filter the data extracting 3 columns: ['Segmento', 'Region', 'Valor_Venda']. Following, we group by two columns: ['Segmento', 'Region']. And, hen we aggreagate the data calculating the average, standard deviation and the count of elements for the column that was out of the group by, in this case the column Valor_Venda.

The function agg() receives as argument a list of functions for aggregation.

### Split the strings in Pandas DataFrames

With Pandas we can realize several split tasks spliting a column or extracting elements of our interest. Let's see the example.

We saw until here several Pandas' features that make the process of data manipulation really simple. And to finish this chapter, let's study the options that Pandas offer to direct chart creation from dataframes, without necessarily use any other library.

```python
# Let's start importing the iris datased from Scikit-learn
from sklearn.datasets import load_iris
data = load_iris()
```

```python
# Then we load iris dataset as a Pandas dataframe
import pandas as pd
df = pd.DataFrame(data['data'], columns = data['feature_names'])
df['species'] = data['target']
df.head()
```

```python
# To create a line graph with all the variable 
# from the dataframe, it just need to do this:
df.plot()
```

```python
# What about a scatter plot with two variable
df.plot.scatter(x = 'sepal length (cm)', y = 'sepal width (cm)')
```

```python
# And even more complex graph, like an area graphic
# might be create:
columns = ['sepal length (cm)', 'petal length (cm)', 'petal width (cm)', 'sepal width (cm)']
df[columns].plot.area()
```

```python
# Calculating the average of colums grouping by the column
# species and creating a bar chart with the result
df.groupby('species').mean().plot.bar()
```

```python
# Or then, we make the count of classes of the column species
# and plot in a pie graphic
df.groupby('species').count().plot.pie(y = 'sepal length (cm)')
```
