### Building Plots

In Matplotlib, a "plot" refers to a graphical representation of data, typically showing the relationship between two or more variables. It's the visual output generated by Matplotlib's functions, most commonly using the matplotlib.pyplot.plot() function.

Here's a breakdown of what a plot entails in Matplotlib:

Types of Plots:

* Matplotlib supports a wide array of plot types, each suited for different data visualization needs, including:
* Line Plots: Showing trends over time or continuous variables.
* Scatter Plots: Illustrating the correlation between two variables.
* Bar Charts: Comparing discrete categories.
* Histograms: Displaying the distribution of a single variable.
* Pie Charts: Representing proportions of a whole.
* Contour Plots: Visualizing 3D data on a 2D plane.
* 3D Plots: Representing data in three dimensions.

In essence, a plot in Matplotlib is the final visual output that helps in understanding and communicating insights from data.

### Stacked Area Graph

A stacked area chart is a type of area chart that displays the evolution of the magnitude of multiple data series (or groups) over time, where each series is stacked on top of the others. The total height of the stack at any point represents the cumulative value of all the series up to that point. This visualization allows for easy comparison of individual series contributions to the total and how those contributions change over time.

**How it works:**

Multiple area charts are stacked on top of each other, with the baseline of each subsequent area chart resting on the top of the area below it. 

```python
days = [1, 2, 3, 4, 5]
sleep = [7, 8, 6, 77, 7]
eat = [2, 3, 4, 5, 3]
work = [7, 8, 7, 2, 2]
stroll = [8, 5, 7, 8, 13] # to walk => 'passear'
```

```python
plt.stackplot(days, sleep, eat, work, stroll, colors = 
              ['m', 'c', 'r', 'k', 'b'])
plt.show()
```

### Pie Graph

A pie chart, also known as a circle chart, is a type of graph that visually represents data as slices of a circle, where each slice represents a proportion of the whole. It's used to show how different categories of data contribute to a total amount. The entire circle represents 100%, and each slice's size is proportional to its corresponding value or percentage. 

```python
slices = [7, 2, 2, 13]
activities = ['sleep', 'eat', 'stroll', 'work']
colors = ['olive', 'lime', 'violet', 'royalblue']
```

```python
plt.pie(slices, labels=activities, colors=colors, startangle=90,
       shadow=True, explode = (0, 0.2, 0, 0))
plt.show()
```

Pylab is a convenience module within the Matplotlib library in Python. It was designed to provide a MATLAB-like interface for plotting and numerical computation by importing functions from both Matplotlib's pyplot module (for plotting) and the NumPy library (for numerical operations) into a single namespace. 

While Pylab offered a simplified way to perform both computations and visualizations in a single line, its use is now generally discouraged in favor of explicitly importing numpy and matplotlib.pyplot separately. ### Line Graph

A line graph, also known as a line chart, is a type of graph that displays information as a series of data points connected by straight line segments. These graphs are particularly useful for visualizing trends and changes in data over a continuous interval, often time, making them valuable for identifying patterns and relationships. 

```python
# Line Chart

# Data
x = linspace(0, 5, 10)
y = x ** 2

# Creates the figure
fig = plt.figure()

# Defines the axes scale
axes = fig.add_axes([0, 0, 0.8, 0.8])

# Creates the plot
axes.plot(x, y, 'r')

# Labels and Title
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('Line Chart')
```

```python
# Line Chart with 2 figures

# Data
x = linspace(0, 5, 10)
y = x ** 2

# Creates the figure
fig = plt.figure()

# Defines the axes scale
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # main figure axes
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # secondary figure axes

# Main Figure
axes1.plot(x, y, 'r')
axes1.set_xlabel('x')
axes1.set_ylabel('y')
axes1.set_title('Main Figure')

# Secondary Figure
axes2.plot(y, x, 'g')
axes2.set_xlabel('y')
axes2.set_ylabel('x')
axes2.set_title('Secondary Figure')
```

```python
# Line Charts in parallel

# Data
x = linspace(0, 5, 10)
y = x ** 2

# Divide the plot area in two subplots
fig, axes = plt.subplots(nrows = 1, ncols = 2)

# Loop through the axes to create the plot
for ax in axes:
    ax.plot(x, y, 'r')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title("Title")
    
# Adjust the layout
fig.tight_layout()
```

```python
# Line Chart with different scales

# Data
x = linspace(0, 5, 10)
y = x ** 2

# Create subplots
fig, axes = plt.subplots(1, 2, figsize=(10,4))

# Creates plot1
axes[0].plot(x, x**2, x, exp(x))
axes[0].set_title("Default scale")

# Creates plo2
axes[1].plot(x, x**2, x, exp(x))
axes[1].set_yscale("log")
axes[1].set_title("Logarithmic scale (y)")
```

### Histogram

A histogram is a graphical representation of the distribution of numerical data. It displays the frequencies of data points falling within specified ranges or bins. Essentially, it shows how often different values occur within a dataset. 

In histograms, the continuos variable is divided in different ranges of values, known as class interval. The horizontal axis represents the class intervals, while the vertical axis represents the frequency of data observations that falls in each class interval. The class intervals must be chosen properly for the data distribution, and the axes scale must be properly defined.

The histograms are useful to visualize the way and the dispersion of continues data distribution, like the height of a population or the punctuations in a test. They might help to identify the presence of atypical values or outliers in data and provide important insights about the data distribution  

```python
# Histogram (hist)

# Data
n = np.random.randn(100000)

# Create the subplots
fig, axes = plt.subplots(1, 2, figsize=(12,4))

# Plot 1
axes[0].hist(n)
axes[0].set_title("Default Histogram")
axes[0].set_xlim((min(n), max(n)))

# Plot 2
axes[1].hist(n, cumulative = True, bins=50)
axes[1].set_title("Cumulative Histogram")
axes[1].set_xlim((min(n), max(n)));
```
